```python
'Close'（收盘价）：股票的收盘价格。
'Factor'（因子）：与股票相关的因子（可能与股票拆分或公司行动有关）。
'FactorPost'（后因子调整）：与股票相关的后因子调整。
'High'（最高价）：股票在特定时期内的最高价格。
'Limit_Down'（跌停价）：股票的跌停价格（通常是由于价格波动引起的）。
'Limit_Up'（涨停价）：股票的涨停价格（通常是由于价格波动引起的）。
'Low'（最低价）：股票在特定时期内的最低价格。
'Open'（开盘价）：股票的开盘价格。
'Pause'（暂停）：与交易暂停相关的信息（可能是由于市场波动或其他因素引起的）。
'Total_Turnover'（总成交额）：股票的总成交额或交易量。
'Volume'（成交量）：股票的成交量（交易的股票数量）。
```

```python
'Capitalization'（总市值）：股票的总市值，即所有已发行股份的市值总和。
'CirculatingCap'（流通市值）：股票的流通市值，即可自由交易的股份的市值总和。
'CirculatingMarketCap'（流通市值市值）：股票的流通市值市值，与流通市值相关的市值。
'MarketCap'（市值）：股票的市值，即公司的整体市场价值。
'PBRatio'（市净率）：市净率，即股票价格与每股净资产的比率。
'PCFRatio'（市现率）：市现率，即股票价格与每股现金流的比率。
'PSRatio'（市销率）：市销率，即股票价格与每股销售收入的比率。
'PeRatio'（市盈率）：市盈率，即股票价格与每股盈利的比率。
'PeRatioLYR'（静态市盈率）：静态市盈率，通常是指过去一年的盈利与股票价格的比率。
'TurnoverRatio'（换手率）：换手率，即股票的交易量与流通股份的比率，通常用于衡量市场活跃度。
```
